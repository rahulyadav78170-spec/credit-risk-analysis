# Credit Risk Analysis

## 1. Business Understanding

### Project Overview

ABC Bank provides various loan products to customers, including personal, business, education, and auto loans. While lending is a major source of revenue, loan defaults can result in significant financial losses. To improve lending decisions and reduce credit risk, the bank wants to analyze historical loan application data.

As a Data Analyst, the objective of this project is to identify the key factors associated with loan default and provide actionable insights that support data-driven lending decisions.

---

# 2. Business Problem

The bank lacks a clear understanding of which customer characteristics and loan attributes contribute to loan defaults. Without this knowledge, loan approval decisions may expose the bank to unnecessary financial risk.

The challenge is to analyze historical loan data and identify patterns that distinguish customers who are more likely to default from those who successfully repay their loans.

---

# 3. Business Objectives

The objectives of this project are:

* Measure the overall loan default rate.
* Identify customer segments with higher default risk.
* Analyze the relationship between customer characteristics and loan default.
* Evaluate the impact of loan attributes on repayment.
* Build an interactive dashboard for business stakeholders.
* Provide recommendations to improve loan approval strategies and reduce financial risk.

# 5. Business Questions

## Portfolio Performance

1. How many loan applications are included in the dataset?
2. What is the overall loan default rate?
3. What is the total loan amount issued?
4. What is the average loan amount?
5. What is the average interest rate?

## Customer Analysis

6. Which age group has the highest default rate?
7. Does income influence loan default?
8. How does credit score affect loan default?
9. Does employment duration influence repayment?
10. Which education level has the highest default rate?
11. Which employment type has the highest default rate?
12. Does marital status affect loan default?

## Loan Analysis

13. Which loan purposes have the highest default rates?
14. Are larger loans associated with higher default risk?
15. Which loan terms experience the highest default rates?
16. Does interest rate influence default?

## Risk Analysis

17. Does having a mortgage reduce default risk?
18. Does having dependents influence default?
19. Does having a co-signer reduce default risk?
20. Which customer profile represents the highest lending risk?
