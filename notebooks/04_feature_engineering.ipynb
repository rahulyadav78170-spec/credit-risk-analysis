```python
import pandas as pd
```

```python
df=pd.read_csv("C:/Users/RAHUL YADAV/OneDrive/credit-risk-analysis/data/cleaned/loan_default_cleaned.csv")
```

```python
age_bins=[18,30,45,60,70]
age_labels=['18-30','31-45','45-60','60+']
df['agegroup']= pd.cut(df['age'],bins=age_bins,labels=age_labels,include_lowest=True)
```

```python
df['agegroup'].unique()
```

```python
df["agegroup"].value_counts()
```

```python
income_bins = [15000, 50000, 100000, 150000]

income_labels = ["Low","Medium","High"]

df["incomegroup"] = pd.cut(df["income"],bins=income_bins,labels=income_labels,include_lowest=True)
```

```python
df["incomegroup"].value_counts()
```

```python
credit_bins = [300,580,670,740,850]

credit_labels = ["Poor","Fair","Good","Excellent"]

df["creditcategory"] = pd.cut(df["creditscore"],bins=credit_bins,labels=credit_labels,include_lowest=True)
```

```python
df["creditcategory"].value_counts()
```

```python
employment_bins = [0,24,60,120]

employment_labels = ["New","Experienced","Senior"]

df["employmentlevel"] = pd.cut(df["monthsemployed"],bins=employment_bins,labels=employment_labels,include_lowest=True)
```

```python
df["employmentlevel"].value_counts()
```
